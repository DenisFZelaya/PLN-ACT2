### 3.2 Verifica la instalación de TensorFlow y la disponibilidad de la GPU

```python
import tensorflow as tf

# Listar GPUs disponibles
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# Verificar el nombre de la GPU
for device in tf.config.list_physical_devices('GPU'):
    print(device.name)
```

### 3.3 Limitar el uso de memoria de la GPU
Configura un límite de memoria fijo para cada GPU. 
Por ejemplo, para limitar el uso de memoria a 4 GB:

> En caso que no funcione en nuestro notebook podemos ejecutar este comando directamente en nuestro notebook.

```python
import tensorflow as tf

# Listar GPUs disponibles
gpus = tf.config.list_physical_devices('GPU')

if gpus:
    try:
        # Limitar el uso de memoria a 4096 MB (4 GB) para la primera GPU
        tf.config.experimental.set_virtual_device_configuration(
            gpus[0],
            [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=4096)]
        )
    except RuntimeError as e:
        print(e)
```
